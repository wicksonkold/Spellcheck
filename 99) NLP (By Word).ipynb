```python
import docx
import contextualSpellCheck
import spacy
from docx import Document
```

```python
nlp = spacy.load("en_core_web_trf")
```

```python
nlp.pipe_names
```

```python
nlp.add_pipe("contextual spellchecker")
```

```python
nlp.pipe_names
```

```python
#Source doc
doc_name = 'GLI_19890303_A(elDoc)'
doc = docx.Document(doc_name + '.docx')


final_temp_list = []
temp_ps = ''

#create new document
newdoc = docx.Document()

#loop the doc and concat them together
for i in range(len(doc.paragraphs)):
    temp_p = doc.paragraphs[i].text
    temp_ps = temp_ps + " " + temp_p
  
#split all words
temp_split = temp_ps.split()

#loop the doc and 
for i in temp_split:
    word = nlp(i)
    check = word._.outcome_spellCheck
    print(word,'  /  ',check)


```

```python
ff = ' '.join(final_temp_list)
print(ff)
```

```python
#save new document
newdoc.add_paragraph(final_temp_list)
newdoc.save(doc_name+'_ED(NLP-by word).docx')
```
